```python
ì# Importa la libreria per creare grafici
import matplotlib.pyplot as plt  # gli do io un nome (plt)
# Crea una lista di animali e una del loro numero corrispondente
animali = ['Elefanti', 'Leoni', 'Tigri', 'Scimmie', 'Zebre']  # lista delle specie animali
numero_animali = [4, 3, 2, 8, 5]  # lista del numero di esemplari per ciascuna specie
# Crea un grafico a barre utilizzando le liste precedenti
plt.bar(animali, numero_animali, color="blue")  # disegna un grafico a barre con le categorie di animali e altezza data da numero_animali, colorando le barre di blu
# Mostra il grafico
plt.show()  # visualizza il grafico
```

```python
# Importa la libreria per creare grafici
import matplotlib.pyplot as plt  

# Definisce un dizionario con gli animali e il numero corrispondente
# La chiave è il tipo di animale e il valore è il loro numero
numero_animali = {'Elefanti': 4, 'Leoni': 3, 'Tigri': 2, 'Scimmie': 8, 'Zebre': 5}  

# Crea un grafico a barre utilizzando le chiavi e i valori del dizionario
# numero_animali.keys() restituisce una vista delle chiavi del dizionario (i nomi degli animali)
# numero_animali.values() restituisce una vista dei valori del dizionario (il numero di ciascun animale)
# le barre sono colorate di rosso
plt.bar(numero_animali.keys(), numero_animali.values(), color="red")  

# Mostra il grafico
plt.show()  
```

```python
# Importa il modulo pyplot dalla libreria matplotlib e rinominalo come plt per facilitarne l'uso
import matplotlib.pyplot as plt 

# Definisci due liste, una con i nomi degli animali e una con il numero corrispondente di ogni animale
animali = ['Elefanti', 'Leoni', 'Tigri', 'Scimmie', 'Zebre']  
numero_animali = [4, 3, 2, 8, 5]  

# Creo un grafico a barre utilizzando le due liste e colora le barre di verde
plt.bar(animali, numero_animali, color="green")  

# Aggiungo un titolo al grafico
plt.title('Numero di animali in uno zoo')  

# Aggiungo un'etichetta per l'asse delle x
plt.xlabel('Animali')  

# Aggiungo un'etichetta per l'asse delle y
plt.ylabel('Numero')  

# Visualizza il grafico
plt.show()
```

```python
# Importa la libreria matplotlib.pyplot con l'alias plt
import matplotlib.pyplot as plt

# Crea due liste: una contenente i mesi e una le corrispondenti temperature medie
mese = ['Gennaio', 'Febbraio', 'Marzo']
temperatura_media = [10, 12, 15]

# Crea un grafico di linea con marker a forma di quadrato ('s'), linea tratteggiata ('--') e colore rosa
# I punti (mesi e temperature) sono collegati da linee tratteggiate e ogni punto è segnato con un quadrato
plt.plot(mese, temperatura_media, marker='s', linestyle='--', color="pink")

# Aggiunge un titolo al grafico
plt.title('Temperatura media mensile')

# Aggiunge etichette agli assi x e y
plt.xlabel('Mese')
plt.ylabel('Temperatura')

# Aggiunge una griglia al grafico per migliorarne la leggibilità
plt.grid(True)

# Visualizza il grafico
plt.show()
```

```python
# Importa il modulo pyplot dalla libreria matplotlib con l'alias plt
import matplotlib.pyplot as plt

# Definisci una lista di colori e due liste con i mesi e le corrispondenti temperature medie
colors = ['gold', 'pink', 'red', 'blue']
mese = ['Gennaio', 'Febbraio', 'Marzo', 'Aprile']
temperatura_media = [10, 15, 20, 35]

# Crea un grafico a torta utilizzando la temperatura_media come valori delle fette,
# mese come etichette delle fette e colors per colorare le fette
plt.pie(temperatura_media, labels=mese, colors=colors)

# Mostra il grafico
plt.show()
```

```python
# Importa le librerie per la creazione di grafici e la manipolazione di dati
import matplotlib.pyplot as plt
import pandas as pd
# Dati sugli studenti e i loro punteggi
nomi_studenti =  ['Alberto', "Laura", 'Giovanni']  # elenco dei nomi degli studenti
punteggi = [13, 22, 40]  # punteggi corrispondenti
# Creazione di un DataFrame con i dati
data = {'Nome dello Studente': nomi_studenti, 'Punteggio': punteggi}  # creazione di un dizionario con i dati
df = pd.DataFrame(data)  # conversione del dizionario in un DataFrame di pandas
# Ordina il DataFrame in base ai punteggi
df.sort_values(by='Punteggio', inplace=True)  # ordina il DataFrame in base al punteggio, modificando il DataFrame originale
# Crea un grafico a barre orizzontali con i dati ordinati
plt.barh(df['Nome dello Studente'], df['Punteggio'], color='green')  # disegna un grafico a barre orizzontali usando i nomi degli studenti come etichette sull'asse y e i loro punteggi sull'asse x, colora le barre di verde
# Imposta il titolo e le etichette degli assi del grafico
plt.title('Punteggi Studenti')  # aggiunge un titolo al grafico
plt.xlabel('Punteggio')  # etichetta per l'asse x (Punteggio)
plt.ylabel('Nome Studente')  # etichetta per l'asse y (Nome dello Studente)
# Mostra il grafico
plt.show()  # visualizza il grafico creato
```

```python
# Importa le librerie necessarie
import matplotlib.pyplot as plt
import pandas as pd
# Dati iniziali sugli studenti e i loro punteggi
nomi_studenti =  ['Alberto', "Laura", 'Giovanni']  # elenco dei nomi degli studenti
punteggi = [13, 22, 40]  # punteggi corrispondenti
# Creazione del DataFrame con i dati
data = {'Nome dello Studente': nomi_studenti, 'Punteggio': punteggi}
df = pd.DataFrame(data)
# Ordina il DataFrame in base ai punteggi, senza sovrascrivere quello originale
df1 = df.sort_values(by='Punteggio', inplace=False)  # crea un nuovo DataFrame ordinato in base ai punteggi
# Crea un grafico a barre orizzontali con i dati ordinati
plt.barh(df1['Nome dello Studente'], df1['Punteggio'], color='green')  # usa i dati dal DataFrame ordinato
# Aggiungi titoli e etichette
plt.title('Punteggi Studenti')  # titolo del grafico
plt.xlabel('Punteggio')  # etichetta dell'asse x
plt.ylabel('Nome Studente')  # etichetta dell'asse y
# Mostra il grafico
plt.show()  # visualizza il grafico
```

```python
import numpy as np  
import matplotlib.pyplot as plt  
# Genera dati casuali per altezza e peso dei maschi e delle femmine usando la distribuzione normale.(media, distribuzione standard, numero di elementi da estrarre)
altezza_maschi = np.random.normal(175, 10, 50) 
peso_maschi = np.random.normal(70, 5, 50)       
altezza_femmine = np.random.normal(162, 8, 50)  
peso_femmine = np.random.normal(60, 5, 50)      
# Crea un grafico scatter per i maschi e le femmine.
# I maschi sono rappresentati con cerchi blu e le femmine con quadrati rosa
plt.scatter(altezza_maschi, peso_maschi, color='blue', label='Maschi', marker='o')  
plt.scatter(altezza_femmine, peso_femmine, color='pink', label='Femmine', marker='s')  
# Aggiungo titoli al grafico e alle assi
plt.title('Altezza e Peso')  # Titolo del grafico
plt.xlabel('Altezza [cm]')   # Etichetta asse X
plt.ylabel('Peso [kg]')      # Etichetta asse Y
# Aggiungi la legenda in alto a destra e una griglia per facilitare la lettura del grafico
plt.legend(loc='upper right')  # Posizione della legenda
plt.grid(True)# Aggiungi la griglia
# Mostra il grafico
plt.show()  
```

```python
# Importa la libreria per creare grafici
import matplotlib.pyplot as plt
# Dati sulle materie e sul numero di studenti maschi e femmine
materie = ["Matematica", "Italiano", "Inglese"]  # elenco delle materie
maschi = [10, 12, 8]  # numero di studenti maschi per materia
femmine = [12, 15, 9]  # numero di studenti femmine per materia
# Imposta la dimensione del grafico
plt.figure(figsize=(10, 6))  # imposta le dimensioni del grafico (larghezza x altezza in pollici)
# Crea un grafico a barre per i maschi e le femmine
width = 0.35  # larghezza delle barre
ind = range(len(materie))  # indici delle materie
# Crea barre per maschi e femmine
plt.bar(ind, maschi, width, label='Maschi', color='blue', alpha=0.7) 
plt.bar([i + width for i in ind], femmine, width, label='Femmine', color='pink', alpha=0.7)
# Aggiungi titoli e etichette
plt.xlabel('Materie')
plt.ylabel('Numero di studenti')
plt.title('Numero di maschi e femmine per materia')
plt.xticks([i + width / 2 for i in ind], materie)  # etichette sull'asse x posizionate al centro tra le barre
# Aggiungi legenda
plt.legend()
# Mostra il grafico
plt.show()
```

```python
# Importa la libreria per creare grafici
import matplotlib.pyplot as plt
# Definizione dei dizionari per i maschi e le femmine con le materie come chiavi
maschi = {'Matematica': 10, 'Italiano': 12, 'Inglese': 8}  # dati dei maschi per materia
femmine = {'Matematica': 12, 'Italiano': 15, 'Inglese': 9}  # dati delle femmine per materia
# Imposta la dimensione del grafico
plt.figure(figsize=(10, 6))  # imposta le dimensioni del grafico (larghezza x altezza in pollici)
# Crea barre per i maschi
plt.bar(maschi.keys(), maschi.values(), label='Maschi', color='blue', alpha=0.7)  # grafico a barre per i maschi
# Crea barre per le femmine, posizionate sopra quelle dei maschi
plt.bar(femmine.keys(), femmine.values(), label='Femmine', color='pink', alpha=0.7, bottom=list(maschi.values()))  # grafico a barre per le femmine, posizionate sopra quelle dei maschi
# Aggiungi titoli e etichette
plt.xlabel('Materie')  # etichetta dell'asse x
plt.ylabel('Numero di studenti')  # etichetta dell'asse y
plt.title('Numero di maschi e femmine per materia')  # titolo del grafico
# Aggiungi legenda
plt.legend()  # mostra la legenda del grafico
# Ottimizza la disposizione degli elementi nel grafico
plt.tight_layout()  # ottimizza la disposizione per evitare sovrapposizioni di elementi
# Mostra il grafico
plt.show()  # visualizza il grafico
```

```python
# Importa la libreria per creare grafici
import matplotlib.pyplot as plt
# Dati degli anni e dei tre gruppi
annata =  ["2020", "2021", "2022", "2023"]  # esempio di anni
gruppo1 = [10, 15, 9, 14]  # dati per il Gruppo 1
gruppo2 = [8, 17, 7, 13]   # dati per il Gruppo 2
gruppo3 = [12, 14, 11, 16] # dati per il Gruppo 3
# Imposta la dimensione del grafico
plt.figure(figsize=(10, 6))  # imposta le dimensioni del grafico (larghezza x altezza in pollici)
# Crea le linee del grafico per ogni gruppo
plt.plot(annata, gruppo1, marker='o', label='Gruppo1', linestyle='-', color='blue')  # linea con cerchi per il Gruppo 1
plt.plot(annata, gruppo2, marker='s', label='Gruppo2', linestyle='--', color='green') # linea tratteggiata con quadrati per il Gruppo 2
plt.plot(annata, gruppo3, marker='^', label='Gruppo3', linestyle='-.', color='red')   # linea punteggiata con triangoli per il Gruppo 3
# Aggiungi titoli e etichette
plt.xlabel('Annata')  # etichetta dell'asse x
plt.ylabel('Valore')  # etichetta dell'asse y
plt.title('Valore per gruppo per annata')  # titolo del grafico
# Aggiungi legenda
plt.legend()  # mostra la legenda per identificare i gruppi
# Abilita la griglia sul grafico
plt.grid(True)  # rende il grafico più leggibile grazie alla griglia
# Mostra il grafico
plt.show()  # visualizza il grafico
```

```python
# Importa le librerie necessarie per la creazione di grafici e manipolazione dei dati
import matplotlib.pyplot as plt
import numpy as np
# Dati degli anni e dei tre gruppi
annata = ["2020", "2021", "2022", "2023"]  # esempio di anni
gruppo1 = [10, 15, 9, 14]  # dati per il Gruppo 1
gruppo2 = [8, 17, 7, 13]   # dati per il Gruppo 2
gruppo3 = [12, 14, 11, 16] # dati per il Gruppo 3
# Impostazioni per le barre
larghezza_barre = 0.2  # larghezza di ciascuna barra
indici = np.arange(len(annata))  # indici numerici per le posizioni delle barre
# Crea un grafico a barre per ciascun gruppo
plt.bar(indici, gruppo1, width=larghezza_barre, label='Gruppo1', color='blue')  # barre per il Gruppo 1
plt.bar(indici + larghezza_barre, gruppo2, width=larghezza_barre, label='Gruppo2', color='green')  # barre per il Gruppo 2
plt.bar(indici + 2*larghezza_barre, gruppo3, width=larghezza_barre, label='Gruppo3', color='red')  # barre per il Gruppo 3
# Aggiungi titoli e etichette
plt.xlabel('Annata')  # etichetta dell'asse x
plt.ylabel('Valore')  # etichetta dell'asse y
plt.title('Valore per gruppo per annata')  # titolo del grafico
# Aggiungi legenda
plt.legend()  # mostra la legenda per identificare i gruppi
# Abilita la griglia sul grafico
plt.grid(True, axis='y')  # rende il grafico più leggibile grazie alla griglia sull'asse y
# Mostra il grafico
plt.show()  # visualizza il grafico
```
