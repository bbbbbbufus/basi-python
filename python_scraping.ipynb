```python
import requests
from bs4 import BeautifulSoup

# Sostituisci questa URL con l'indirizzo del sito che vuoi analizzare
url = "https://www.example.com"

# Effettua una richiesta GET alla pagina web
response = requests.get(url)

# Analizza il contenuto HTML della pagina utilizzando BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Estrai e stampa i titoli dalla pagina
for heading in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6']):
    print(heading.text.strip())
    
    
for parag in soup.find_all("p"):
    print(parag.text.strip())
```

```python
# trova tutti i link
for link in soup.find_all(['a']):
    print(link.text.strip())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Sostituisci questa URL con l'indirizzo del sito che vuoi analizzare
url = "https://it.wikipedia.org/wiki/Campionato_mondiale_di_Formula_1_2022"

# Effettua una richiesta GET alla pagina web
response = requests.get(url)

# Analizza il contenuto HTML della pagina utilizzando BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Trova tutte le tabelle nella pagina
tables = soup.find_all('table')

# Leggi tutte le tabelle utilizzando pandas
for i, table in enumerate(tables):
    # Utilizziamo try-except per gestire eventuali errori durante l'importazione delle tabelle
    try:
        df = pd.read_html(str(table))[0]
        print(f"Table {i + 1}:\n{df}\n{'='*40}\n")
    except Exception as e:
        print(f"Errore durante l'importazione della tabella {i + 1}: {str(e)}\n{'='*40}\n")
```
