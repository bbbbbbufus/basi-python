```python
# Importiamo le librerie necessarie
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Creiamo un set di dati di esempio con valori mancanti
data = {
    'Variable1': [1, 2, 3, 4, 5],         # Colonna numerica
    'Variable2': [1, 2, np.nan, 4, np.nan], # Colonna numerica con valori mancanti
    'Missing_Column': ['A', 'B', 'A', 'C', np.nan] # Colonna categorica con valori mancanti
}

# Convertiamo i dati in un DataFrame pandas
df = pd.DataFrame(data)
df1 = pd.DataFrame() # Creiamo un nuovo DataFrame vuoto per i dati elaborati

# Selezioniamo le colonne numeriche del DataFrame
numeric_cols = df.select_dtypes(include=['number'])
# Riempiamo i valori NaN nelle colonne numeriche con la media di ciascuna colonna
df1[numeric_cols.columns] = numeric_cols.fillna(numeric_cols.mean())

# Selezioniamo le colonne categoriche del DataFrame
categorical_cols = df.select_dtypes(include=['object'])
# Riempiamo i valori NaN nelle colonne categoriche con la moda (il valore più frequente) di ciascuna colonna
df1[categorical_cols.columns] = categorical_cols.fillna(categorical_cols.mode().iloc[0])

# Stampiamo i DataFrame originali e modificati
print(f"il primo con i valori mancanti\n{df} \ne il secondo con i missing values sostituiti \n{df1}")
```

```python
# Importazione delle librerie necessarie
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generazione di un set di dati di esempio con valori mancanti
data = {
    'Variable1': [1, 2, 3, 4, 5],                  # Colonna numerica
    'Variable2': [1, 2, np.nan, 4, np.nan],        # Colonna numerica con valori NaN
    'Missing_Column': ['A', 'B', 'A', 'C', np.nan] # Colonna categorica con un valore NaN
}

# Creazione di un DataFrame da questi dati
df = pd.DataFrame(data)
df1 = pd.DataFrame() # Creazione di un DataFrame vuoto per i dati elaborati

# Definizione della funzione per trattare i valori mancanti
def missingvalues_sub(df):
    # Selezione delle colonne numeriche e riempimento dei valori NaN con la media
    numeric_cols = df.select_dtypes(include=['number'])
    df1[numeric_cols.columns] = df[numeric_cols.columns].fillna(df[numeric_cols.columns].mean())
    
    # Selezione delle colonne categoriche e riempimento dei valori NaN con la moda
    categorical_cols = df.select_dtypes(exclude=['number'])
    df1[categorical_cols.columns] = df[categorical_cols.columns].fillna(df[categorical_cols.columns].mode().iloc[0])
    
    return df1

# Funzione principale che esegue il programma
def main():
    df1 = missingvalues_sub(df)
    # Stampa dei DataFrame originali e modificati
    print(f"il primo con i valori mancanti \n{df} \ne il secondo con i missing values sostituiti \n{df1}")

# Punto di ingresso del programma
if __name__ == "__main__":
    main()
```

```python
import numpy as np
import seaborn as sns      
import matplotlib.pyplot as plt

# Genera un dataset di esempio con variabili numeriche.
np.random.seed(42)  # Imposta il seed per la generazione di numeri casuali, garantendo la riproducibilità.
data = pd.DataFrame(np.random.rand(100, 5), 
                    columns=['Var1', 'Var2', 'Var3', 'Var4', 'Var5'])  # Crea un DataFrame con 100 righe e 5 colonne di valori casuali tra 0 e 1.

# Aggiungi alcune variabili categoriche generate casualmente.
data['Categoria1'] = np.random.choice(['A', 'B', 'C'], size=100)  # Aggiunge una nuova colonna 'Categoria1', con valori casuali tra 'A', 'B' e 'C'.
data['Categoria2'] = np.random.choice(['X', 'Y'], size=100)  # Aggiunge una nuova colonna 'Categoria2', con valori casuali tra 'X' e 'Y'.

# Calcola la matrice di correlazione delle variabili numeriche nel DataFrame.
correlation_matrix = data.corr()  # Utilizza la funzione 'corr()' per calcolare la correlazione tra le colonne numeriche.

# Imposta la dimensione della figura per il grafico. 
# Il parametro figsize=(larghezza, altezza) definisce le dimensioni della figura in pollici.
plt.figure(figsize=(10, 8))

# Crea un heatmap utilizzando Seaborn. 
# 'correlation_matrix' è la matrice di correlazione che vuoi visualizzare.
sns.heatmap(correlation_matrix, 
            annot=True,  # 'annot=True' aggiunge le etichette (i valori di correlazione) in ciascuna cella del heatmap.
            cmap='coolwarm',  # 'cmap' definisce la palette di colori da usare. 'coolwarm' è una palette da blu (valori bassi) a rosso (valori alti).
            fmt='.2f',  # 'fmt' specifica il formato numerico delle etichette. Qui si usano due cifre decimali.
            alpha=0.7)  # 'alpha' definisce la trasparenza delle celle. 0 è completamente trasparente, 1 è completamente opaco.

# Imposta il titolo del grafico.
plt.title("Matrice di correlazione")

# Mostra il grafico. Questa funzione rende visibile il grafico a schermo.
plt.show()
```

```python
import pandas as pd
import numpy as np

# Imposta il seed per la generazione di numeri casuali per rendere i risultati riproducibili.
np.random.seed(41)

# Crea un DataFrame vuoto.
df = pd.DataFrame()

# Genera dati casuali per il DataFrame.
n_rows = 100  # Numero di righe nel DataFrame.

# Genera una colonna di dati categorici casuali scegliendo tra 'A', 'B', e 'C'.
df['CatCol1'] = np.random.choice(['A', 'B', 'C'], size=n_rows)

# Genera un'altra colonna di dati categorici casuali scegliendo tra 'X' e 'Y'.
df['CatCol2'] = np.random.choice(['X', 'Y'], size=n_rows)

# Genera una colonna di dati numerici da una distribuzione normale standard (media 0, deviazione standard 1).
df['NumCol1'] = np.random.randn(n_rows)

# Genera una colonna di dati numerici interi casuali tra 1 e 99 (100 escluso).
df['NumCol2'] = np.random.randint(1, 100, size=n_rows)

# Genera una colonna di dati numerici da una distribuzione uniforme tra 0 e 1.
df['NumCol3'] = np.random.uniform(0, 1, size=n_rows)

# Calcola il numero totale di valori mancanti desiderati nel DataFrame.
total_missing_values = int(0.03 * n_rows * len(df.columns))  # 3% dei valori totali.

# Introduce valori mancanti in modo casuale nel DataFrame.
for column in df.columns:
    # Determina casualmente quanti valori mancanti inserire in ogni colonna.
    num_missing_values = np.random.randint(0, total_missing_values + 1)
    
    # Seleziona indici casuali in cui inserire i valori mancanti.
    missing_indices = np.random.choice(n_rows, size=num_missing_values, replace=False)
    
    # Imposta i valori mancanti (NaN) in quegli indici per la colonna corrente.
    df.loc[missing_indices, column] = np.nan

# Visualizza il DataFrame finale.
df
```

```python
# Selezione delle colonne numeriche del DataFrame
numeric_cols = df.select_dtypes(include=['number'])

# Selezione delle colonne categoriche del DataFrame
categorical_cols = df.select_dtypes(exclude=['number'])

# Sostituzione dei missing values nelle colonne categoriche con la moda (valore più frequente)
# '.loc[:, categorical_cols.columns]' si riferisce a tutte le righe e alle colonne categoriche specificate
# 'fillna' con 'mode().iloc[0]' sostituisce NaN con la moda per ogni colonna categorica
df.loc[:, categorical_cols.columns] = df[categorical_cols.columns].fillna(df[categorical_cols.columns].mode().iloc[0])

# Calcolo della media condizionata per le colonne numeriche, raggruppando per 'CatCol1'
# 'groupby('CatCol1')[numeric_cols.columns].transform('mean')' calcola la media per ogni gruppo in 'CatCol1'
conditional_means = df[numeric_cols.columns].fillna(df.groupby('CatCol1')[numeric_cols.columns].transform('mean'))

# Aggiornamento delle colonne numeriche con la media condizionata
# Utilizza '.loc' per sostituire i valori in tutte le righe e colonne numeriche
df.loc[:, numeric_cols.columns] = conditional_means

# Stampa del DataFrame dopo la gestione dei missing values
print(df)
```
