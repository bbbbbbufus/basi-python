    CONTROLLO SE UN NUMERO è PARI O DISPARI
</h4>

```python
# Stampa un messaggio per chiedere all'utente di inserire un numero
print("Numero: ")
num1 = float(input())  # Legge il numero inserito dall'utente e lo converte in un numero decimale
# Controlla se il numero è pari o dispari
if num1 % 2 == 0:
    # Se il resto della divisione del numero per 2 è 0, allora il numero è pari
    print("è pari")
else:
    # Se il resto della divisione del numero per 2 non è 0, allora il numero è dispari
    print("è dispari")
```

<h4 align="center" style="font-size:30px; font-family:'Arial', sans-serif; color: #e76f51; text-shadow: 1px 1px 3px #f4a261; letter-spacing: 0.5px; margin-top: 15px; margin-bottom: 15px; border-bottom: 2px solid #264653; padding-bottom: 5px;">
    AREA E PERIMETRO DEL QUADRATO
</h4>

```python
# Stampa un messaggio per chiedere all'utente di inserire la lunghezza del lato del quadrato
print("lato: ")
lato = float(input())  # Legge la lunghezza del lato inserita dall'utente e la converte in un numero decimale
# Calcola il perimetro del quadrato
print("calcola il perimetro: ")
perimetro = lato * 4  # Il perimetro di un quadrato è uguale alla lunghezza del lato moltiplicata per 4
# Stampa il valore del perimetro del quadrato
print("il perimetro è: ")
print(perimetro)
# Calcola l'area del quadrato
print("calcola l'area: ")
area = lato * lato  # L'area di un quadrato è uguale alla lunghezza del lato elevata al quadrato
# Stampa il valore dell'area del quadrato
print("l'area è: ")
print(area)
```

<h4 align="center" style="font-size:30px; font-family:'Arial', sans-serif; color: #e76f51; text-shadow: 1px 1px 3px #f4a261; letter-spacing: 0.5px; margin-top: 15px; margin-bottom: 15px; border-bottom: 2px solid #264653; padding-bottom: 5px;">
    CALCOLO AREA E CIRCONFERENZA DEL CERCHIO
</h4>

```python
# Stampa un messaggio per chiedere all'utente di inserire la lunghezza del raggio del cerchio
print("raggio: ")
raggio = float(input())  # Legge la lunghezza del raggio inserita dall'utente e la converte in un numero decimale
# Stampa un messaggio per chiedere all'utente di inserire il valore di pigreco
print("pigreco: ")
pig = float(input())  # Legge il valore di pigreco inserito dall'utente e lo converte in un numero decimale
# Calcola l'area del cerchio
print("l'area è: ")
area = pig * raggio ** 2  # L'area di un cerchio è uguale a pigreco moltiplicato per il raggio al quadrato
# Stampa il valore dell'area del cerchio
print(area)
# Calcola la circonferenza del cerchio
print("la circonferenza: ")
circonferenza = raggio * 2 * pig  # La circonferenza di un cerchio è uguale al diametro (raggio * 2) moltiplicato per pigreco
# Stampa il valore della circonferenza del cerchio
print(circonferenza)
```

```python
# Stampa un messaggio per chiedere all'utente di inserire il valore della base minore del trapezio
print("base minore: ")
baseminore = float(input())  # Legge il valore della base minore inserito dall'utente e lo converte in un numero decimale
# Stampa un messaggio per chiedere all'utente di inserire il valore della base maggiore del trapezio
print("base maggiore: ")
basemaggiore = float(input())  # Legge il valore della base maggiore inserito dall'utente e lo converte in un numero decimale
# Stampa un messaggio per chiedere all'utente di inserire il valore dell'altezza del trapezio
print("altezza: ")
altezza = float(input())  # Legge il valore dell'altezza inserito dall'utente e lo converte in un numero decimale
# Stampa un messaggio per annunciare il calcolo dell'area
print("area è uguale a: ")
# Corregge la formula per calcolare l'area del trapezio
area = (baseminore + basemaggiore) * altezza / 2
# Stampa il valore dell'area del trapezio
print(area)
```

    SEMPLICE CALCOLATRICE
</h4>

```python
# Stampa un messaggio che chiede all'utente di scegliere un'operazione matematica
print("Che operazione vuoi fare? (differenza/addizione/divisione/moltiplicazione): ")
operazione = input()  # Legge l'operazione scelta dall'utente
# Stampa un messaggio che chiede all'utente di inserire il primo numero
print("Inserisci il primo numero: ")
num1 = float(input())  # Legge il primo numero e lo converte in un numero decimale
# Stampa un messaggio che chiede all'utente di inserire il secondo numero
print("Inserisci il secondo numero: ")
num2 = float(input())  # Legge il secondo numero e lo converte in un numero decimale
# Esegue l'operazione scelta in base all'input dell'utente
if operazione == "differenza":
    op = num1 - num2  # Calcola la differenza tra i due numeri
else:
    if operazione == "addizione":
        op = num1 + num2  # Calcola la somma dei due numeri
    else:
        if operazione == "divisione":
            op = num1 / num2  # Calcola la divisione del primo numero per il secondo
        else:
            if operazione == "moltiplicazione":
                op = num1 * num2  # Calcola il prodotto dei due numeri
# Stampa il risultato dell'operazione
print("Il risultato è: ")
print(op)
```

<h4 align="center" style="font-size:30px; font-family:'Arial', sans-serif; color: #e76f51; text-shadow: 1px 1px 3px #f4a261; letter-spacing: 0.5px; margin-top: 15px; margin-bottom: 15px; border-bottom: 2px solid #264653; padding-bottom: 5px;">
    CALCOLO DEL FATTORIALE E DEL QUADRATO DI 5 NUMERI CON CICLO FOR
</h4>

```python
# Inizializza una lista di 5 elementi, tutti impostati a 0
numeri = [0] * 5
# Imposta ciascun elemento della lista 'numeri' a 0
# Questi passaggi sono ridondanti dato che la lista è già stata inizializzata a zero, pero per evitare errori è meglio rispecificare.
numeri[0] = 0
numeri[1] = 0
numeri[2] = 0
numeri[3] = 0
numeri[4] = 0
# Inizializza la variabile 'fattoriale' a 1
fattoriale = 1
# Chiede all'utente di inserire cinque numeri
print("Dammi dei numeri: ")
for i in range(5):  # Itera 5 volte per ottenere 5 numeri dall'utente
    numeri[i] = int(input())  # Legge un numero intero dall'utente e lo assegna all'elemento corrispondente nella lista
# Calcola e stampa i fattoriali dei numeri inseriti
print("I fattoriali sono: ")
for i in range(5):  # Itera attraverso ciascun numero nella lista
    fattoriale = 1
    for k in range(1, numeri[i] + 1):  # Calcola il fattoriale del numero
        fattoriale *= k
    print(fattoriale)  # Stampa il fattoriale del numero
# Stampa i quadrati dei numeri inseriti
print("I numeri al quadrato sono: ")
for i in range(5):  # Itera attraverso ciascun numero nella lista
    print(numeri[i] ** 2)  # Calcola e stampa il quadrato del numero
```

```python
# Stampa un messaggio che chiede all'utente di inserire la quantità di dollari che possiede
print("Quanti dollari hai?: ")
dollari = float(input())  # Legge il valore inserito dall'utente e lo converte in un numero decimale
# Stampa un messaggio che chiede all'utente di inserire la quantità di sterline che possiede
print("Quante sterline hai?: ")
sterline = float(input())  # Legge il valore inserito dall'utente e lo converte in un numero decimale
# Stampa un messaggio che chiede all'utente di inserire il tasso di conversione attuale per i dollari
print("Tasso di Conversione dollari: ")
dollaroconv = float(input())  # Legge il tasso di conversione per i dollari e lo converte in un numero decimale
# Stampa un messaggio che chiede all'utente di inserire il tasso di conversione attuale per le sterline
print("Tasso di Conversione Sterline: ")
sterlinaconv = float(input())  # Legge il tasso di conversione per le sterline e lo converte in un numero intero
# Stampa un messaggio per annunciare il totale in euro
print("Euro totali: ")
# Calcola il totale in euro convertendo sia i dollari che le sterline usando i rispettivi tassi di conversione
eurototali = dollari * dollaroconv + sterline * sterlinaconv
# Stampa il totale in euro
print(eurototali)
```
